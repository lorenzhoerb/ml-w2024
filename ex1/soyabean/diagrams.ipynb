```python
import matplotlib.pyplot as plt
import numpy as np

# Data for Average Scoring
classifiers = ["Bayers", "Decision Tree", "KNN"]
metrics = ["Accuracy", "Precision", "Recall", "F1"]
average_scores = [
    [96, 98, 98, 98],  # Bayers
    [90, 95, 94, 95],  # Decision Tree
    [91, 96, 91, 92]   # KNN
]

# Data for F1 Scoring
f1_scores = [
    [96, 98, 98, 98],  # Bayers
    [91, 95, 94, 94],  # Decision Tree
    [91, 96, 91, 92]   # KNN
]

# Bar chart configuration
x = np.arange(len(metrics))  # Label locations for metrics
width = 0.35  # Width of the bars
num_classifiers = len(classifiers)

# Calculate positions for grouped bars
x_positions = np.arange(len(metrics) * num_classifiers) / num_classifiers * (1.5 * len(metrics))

fig, ax = plt.subplots(figsize=(12, 6))

# Loop through classifiers to create groups of bars
for i, classifier in enumerate(classifiers):
    offset = i * len(metrics)  # Offset groups per classifier
    ax.bar(
        x_positions[offset:offset + len(metrics)] - width / 2,
        average_scores[i],
        width,
        label=f'{classifier} (Avg Scoring)',
        color=f'C{i}'
    )
    ax.bar(
        x_positions[offset:offset + len(metrics)] + width / 2,
        f1_scores[i],
        width,
        label=f'{classifier} (F1 Scoring)',
        color=f'C{i}',
        hatch='//'
    )

# Add some text for labels, title, and custom x-axis tick labels, etc.
ax.set_xlabel('Metrics')
ax.set_ylabel('Scores [%]')
ax.set_title('Classifier Performance Metrics')
ax.set_xticks(x_positions[::len(metrics)] + (len(metrics) - 1) / 2)
ax.set_xticklabels(classifiers)
ax.legend()

# Display the chart
plt.tight_layout()
plt.show()
```
